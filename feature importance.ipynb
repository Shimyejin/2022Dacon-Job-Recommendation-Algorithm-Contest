```python
import pandas as pd
import numpy as np
import numpy as np
import random
import os
import warnings
from glob import glob
from tqdm import tqdm
from sklearn.metrics import f1_score
```

```python
def seed_everything(seed: int = 123456):
    random.seed(seed)
    np.random.seed(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)
    
warnings.filterwarnings(action='ignore')
seed_everything(123456)
```

```python
#path = "C:\\Users\\User\\Desktop\\데이콘\\KNOW_data\\train"
path = "train"

# 파일명을 담을 리스트 생성
file_list = []

# csv로 끝나는 파일명 다 가져오기
for x in glob(path + '/*.csv'):
    file_list.append(x)
file_list.sort()

# 전체 데이터를 담을 리스트 생성
know_train = []

for file in file_list :
    temp = pd.read_csv(file)
    know_train.append(temp)
```

```python
#path_test = "C:\\Users\\User\\Desktop\\데이콘\\KNOW_data\\test"
path_test = "test"

# 파일명을 담을 리스트 생성
file_list_test = []

# csv로 끝나는 파일명 다 가져오기
for x in glob(path_test + '/*.csv'):
    file_list_test.append(x)
file_list_test.sort()

# 전체 데이터를 담을 리스트 생성
know_test = []

for file in file_list_test :
    temp = pd.read_csv(file)
    know_test.append(temp)
```

# 라벨 인코딩

```python
from sklearn.preprocessing import LabelEncoder
years = ['2017', '2018', '2019', '2020']

year_encoder = {}

for year, df in zip(years, know_train):
    print(year)
    encoders = {}
    
    for col in df.columns:
        if col == 'idx':
            continue
        
        try:
            df[col] = df[col].map(int)
        except:
            encoder = LabelEncoder()
            df[col] = df[col].map(str)
            df[col] = encoder.fit_transform(df[col])
            encoders[col] = encoder
            
            
    year_encoder[year] = encoders
```
